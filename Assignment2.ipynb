# Nominal Scale

A nominal scale is a scale (of measurement) in which situations (measurements) are categorized into classes using labels. Factors such as gender, place of residence, types of fruit, and so on are examples of nominal scale variables.  This measurement normally deals only with non-numeric (quantitative) variables or where numbers have no value.

```python
import matplotlib.pyplot as plt

# Sample Gender Data
gender_data = ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male', 'Male', 'Male']

# Count the frequency of each gender
gender_count = {'Male': gender_data.count('Male'), 'Female': gender_data.count('Female')}

# Labels and sizes for the pie chart
labels = list(gender_count.keys())
sizes = list(gender_count.values())

# Create the pie chart
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])

# Equal aspect ratio ensures the pie is drawn as a circle.
plt.axis('equal')
plt.title('Gender Distribution')
plt.show()
```

A pie chart is an appropriate visualization for nominal data because it effectively represents distinct, unordered categories such as gender, fruit types, or car brands. It is designed to show how a whole is divided into different categories, making it ideal for illustrating proportions or percentages. This is particularly useful when the goal is to understand the distribution of each category relative to the total, such as the percentage of males versus females in a group. The pie chartâs clear visual display highlights the relative sizes of the categories, focusing on proportions rather than any inherent order, which suits the nature of nominal data. Additionally, the use of different colors or segments helps distinguish between categories without needing numerical values or ordered information.

# Ordinal Scale

The second level of measurement, known as the ordinal scale, provides the data's ranking and ordering without really determining the degree of variance among them.      

âOrdinalâ indicates âorderâ. Ordinal data is quantitative data which have naturally occurring orders and the difference between is unknown. It can be named, grouped,customer satisfaction ratings, education levels, ranks in a competition, etc.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample Customer Satisfaction Data (Ordinal)
ordinal_data = [5, 4, 3, 4, 2, 5, 1, 3, 4, 2, 3, 3, 1, 2,]

# Count the frequency of each satisfaction level
satisfaction_count = pd.Series(ordinal_data).value_counts().sort_index()

# Create the bar graph
plt.figure(figsize=(8,4))
satisfaction_count.plot(kind='bar', color='blue', edgecolor='black')

# Add labels and title
plt.title('Customer Satisfaction Levels')
plt.xlabel('Satisfaction Level')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

For ordinal data, a bar graph is appropriate for ordinal scale data because it effectively displays ordered categories and their frequencies, preserving the natural ranking of the data. Ordinal data, such as satisfaction levels ("Very Unsatisfied" to "Very Satisfied"), has a meaningful order, and bar graphs clearly show the frequency or count of each ranked category. They allow for easy comparison between categories, helping to visualize how many responses fall into each level, such as the number of customers who rated their satisfaction as 1 versus 5. The bars can be arranged to reflect the hierarchy of the data, emphasizing the distribution and ranking, which aligns well with the ordinal nature of the information. Bar graphs also offer a clear, straightforward interpretation, making it easy to understand the relative frequencies of each category, which is crucial when comparing the magnitude of responses or measurements.


# Interval Scale

 The interval scale has ordered values with meaningful, equal intervals between them,  the variables are measured in an exact manner, not in a relative way in which the presence of zero is arbitrary. The interval scale is positioned above the nominal and ordinal scales in terms of measurement levels. The location parameters mean, median, and mode may all be computed on interval scales.


```python
# Interval Data: Temperature in Celsius
interval_data = pd.DataFrame({
    'Temperature (C)': [25, 30, 28, 35, 22, 23, 32, 29, 27, 24]
})

# Line plot for interval data
interval_data.plot(kind='line', marker='o', color='red')
plt.title('Temperature Readings Over Time')
plt.ylabel('Temperature (Â°C)')
plt.xlabel('Time (Index)')
plt.show()
```

Interval data is best represented by a line plot, particularly when the data shows a time series or continuous change, like temperature it illustrates how temperature data vary over time. This graphical format successfully displays trends or patterns since the data is continuous with equal intervals.

# Ratio

Ratio scales are a particular kind of quantitative variable measuring scale. Comparing the intervals or discrepancies is made possible for any researcher. With a zero point or character of origin, the ratio scale is the fourth level of measurement. This scale has this particular attribute. It is 0 degrees Celsius, for instance, outside. Not that 0 degrees isn't hot or cold, it is a value.

 Ratio scale helps to understand the ultimate-order, interval, values, and the true zero characteristic is an essential factor in calculating ratios. The most common examples of this scale are height, money, age, weight etc. With respect to market research, the common examples that are observed are sales, price, number of customers, market share etc.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample ratio data (weights in kilograms)
data_ratio = pd.DataFrame({
    'Weight (kg)': [70, 65, 80, 55, 64, 75, 66, 72, 85, 91, 78, 67]
})

# Plotting the histogram
plt.hist(data_ratio['Weight (kg)'], bins=5, color='yellowgreen', edgecolor='black')
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()
```

A histogram is often appropriate for visualizing ratio scale data because ratio data is continuous and numeric, meaning values can fall anywhere along a scale. For ratio data, which includes a meaningful zero point (e.g., 0 kg represents no weight), histograms are well-suited as they start the x-axis from zero, accurately reflecting this true zero. Additionally, histograms reveal the distribution patterns of the data, indicating whether it is skewed or follows a normal distribution, which helps in understanding patterns like weight concentration within certain ranges. By summarizing how data is distributed across different ranges, histograms make it easy to see the most common weight ranges and the frequency of various weights. Overall, histograms are valuable for illustrating the spread, distribution, and frequency of ratio scale data, while effectively incorporating its true zero point and continuous nature.
